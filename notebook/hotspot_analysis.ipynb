```python
# Import required libraries
import pandas as pd
import folium
from folium.plugins import HeatMap
```

```python
# Read the Amsterdam city center dataset
df_ams_cc = pd.read_csv('../data/cleaned/df_ams_cc.csv')
print('Dataset shape:', df_ams_cc.shape)
print('\nFirst few rows:')
df_ams_cc.head()
```

```python
# Create base map centered on Amsterdam
amsterdam_map = folium.Map(location=[52.3676, 4.9041],  # Amsterdam center coordinates
                          zoom_start=13,
                          tiles='CartoDB positron'  # Light theme for better visibility
                          )

# Prepare data for heatmap
locations = df_ams_cc[['latitude', 'longitude']].values.tolist()

# Add heatmap layer
HeatMap(locations,
        radius=15,  # Size of each point on the heatmap
        blur=10,    # Amount of blur
        max_zoom=13,  # Maximum zoom level
        gradient={0.4: 'blue', 0.65: 'lime', 1: 'red'}  # Custom gradient
        ).add_to(amsterdam_map)

# Display the map
amsterdam_map
```
